```python
def read_points(pts_file):
    pts = np.genfromtxt(pts_file, delimiter=',', skip_header=3, usecols=(0,1,2,3,8))
    return pts
```

```python
def plot_range_hist(pts, ax, bins=np.arange(0, 100, 0.5)):
    inpts_flag = pts[:, 4] != 0
    hist, bin_edges = np.histogram(pts[inpts_flag, 4], bins=bins)
    ax.bar(bin_edges[:-1], hist)
    return hist, (bin_edges[:-1]+bin_edges[1:])*0.5
```

```python
bins = np.arange(0, 100, 0.5)
swir_hists_2015 = pd.DataFrame(data=np.zeros((len(bins)-1, len(swir_pts_files)+1), dtype=np.float), columns=['bin_center', 'C', 'E', 'N', 'S', 'W'])
for n, fname in enumerate(swir_pts_files):
#     pts = read_points(fname)
    fig, ax = plt.subplots()
    swir_hists_2015.iloc[:, n+1], bin_center = plot_range_hist(read_points(fname), ax, bins=bins)
#     import pdb; pdb.set_trace()
#     swir_hists_2015.iloc[:, n+1] = hist
    ax.set_title(swir_hists_2015.columns.values[n+1])
#     
swir_hists_2015.iloc[:, 0] = bin_center
```

* plot the histograms of range values for 2014 and 2015 together

```python
for plot_name in ('C', 'E', 'N', 'S', 'W'):
    fig, ax = plt.subplots(figsize=(6.5, 5))
    tmpind = np.where(swir_hists_2014.loc[:, 'bin_center'] >= 30)[0]
    tmpind = tmpind[0]
    rg_diff = 30
    rg_res = swir_hists_2014.loc[1, 'bin_center'] - swir_hists_2014.loc[0, 'bin_center']
    ax.bar(swir_hists_2014.loc[tmpind:tmpind+np.fix(rg_diff/rg_res), 'bin_center'], swir_hists_2014.loc[tmpind:tmpind+np.fix(rg_diff/rg_res), plot_name], color='red')
    
    tmpind = np.where(swir_hists_2015.loc[:, 'bin_center'] >= 30)[0]
    tmpind = tmpind[0]
    rg_diff = 30
    rg_res = swir_hists_2015.loc[1, 'bin_center'] - swir_hists_2015.loc[0, 'bin_center']
    ax.bar(swir_hists_2015.loc[tmpind:tmpind+np.fix(rg_diff/rg_res), 'bin_center'], swir_hists_2015.loc[tmpind:tmpind+np.fix(rg_diff/rg_res), plot_name], color='green', alpha=0.5)
    ax.set_title(plot_name)
    ax.set_xlabel('range, meter')
    ax.set_ylabel('frequency')
    ax.legend(['2014', '2015'])
    plt.savefig('./figures/range_hist_2014_vs_2015_zoom_in_30_to_60_{0:s}.png'.format(plot_name), dpi=dpi, bbox_inches='tight', pad_inches=0.)
```
